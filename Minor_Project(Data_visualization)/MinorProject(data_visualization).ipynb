```python
project_path = os.getcwd()
file_path = project_path + '/marvel-wikia-data.csv'
marvel = pd.read_csv(file_path)
```

```python
dc_year = dc.dropna(subset=['YEAR'])
marvel_year = marvel.dropna(subset=['Year'])
```

# Visualization :

### 1.Gender Distribution 

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
dc['SEX'].value_counts().plot(kind='bar')
plt.title("Gender Distribution in DC")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.subplot(1,2,2)
marvel['SEX'].value_counts().plot(kind='bar')
plt.title("Gender Distribution in Marvel")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

### 2.Alignment Distribution (Good / Bad / Neutral)

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
dc['ALIGN'].value_counts().plot(kind='bar')
plt.title("Character Alignment in DC")

plt.subplot(1,2,2)
marvel['ALIGN'].value_counts().plot(kind='bar')
plt.title("Character Alignment in Marvel")

plt.tight_layout()
plt.show()
```

### 3.Alive vs Deceased Characters

```python
plt.figure(figsize=(10,5))

sns.countplot(data=dc, x='ALIVE')
plt.title("Alive vs Deceased Characters in DC")
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(data=marvel, x='ALIVE')
plt.title("Alive vs Deceased Characters in Marvel")
plt.show()
```

### 4.Characters Introduced over the years

```python
plt.figure(figsize=(12,6))

plt.hist(dc_year['YEAR'],
         bins=25,
         alpha=0.7,
         color='crimson',
         edgecolor='black',
         label='DC')

plt.hist(marvel_year['Year'],
         bins=25,
         alpha=0.7,
         color='royalblue',
         edgecolor='black',
         label='Marvel')

plt.title("Characters Introduced Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Characters")
plt.legend()
plt.show()
```

### 5.Top 10 Characters by Appearances

```python
top_dc = dc.sort_values(by='APPEARANCES', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='APPEARANCES', y='name', data=top_dc)
plt.title("Top 10 DC Characters by Appearances")
plt.show()
```

```python
top_marvel = marvel.sort_values(by='APPEARANCES', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='APPEARANCES', y='name', data=top_marvel)
plt.title("Top 10 Marvel Characters by Appearances")
plt.show()
```

### 6.Identity Status Distribution

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
dc['ID'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Identity Status in DC")
plt.ylabel("")

plt.subplot(1,2,2)
marvel['ID'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Identity Status in Marvel")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

### 7.Gender vs Alignment

```python
plt.figure(figsize=(12,6))
sns.countplot(data=marvel, x='ALIGN', hue='SEX')
plt.title("Gender vs Alignment in Marvel Comics")
plt.show()
```
